## 2. Processing Page 3
This is the data about 
Travel miles in the functional classes.
functional Classes are;
      Rural Interstate		
      Rural Other Arterial		
      Other Rural		
      Urban Interstate		
      Urban Other Arterial		
      Other Urban		
      All Systems	
    
 

```python
tvt_p3 = pd.read_excel(file, sheet_name="Page 3", skiprows = 4, skipfooter= 2) # read file
```

```python
tvt_p3
```

```python
tvt_p3.drop(['Unnamed: 1','Unnamed: 2','INDEX'], axis=1, inplace=True) 
#tvt_p3 = tvt_p3.iloc[8:15].reset_index(drop=True)
tvt_p3 = tvt_p3.iloc[8:15]
tvt_p3 = tvt_p3.dropna(axis = 1)

tvt_p3.reset_index(inplace=True)
tvt_p3 = tvt_p3.rename(columns={'Unnamed: 0':'System'})

# reset index
tvt_p3.iloc[:, 1:].reset_index(drop=True)
tvt_p3 =tvt_p3.rename (columns ={'System': "Months"})

# Transpose
tvt_p3 = tvt_p3.T

# get the row index 1 for row name
newheader = tvt_p3.iloc[1]

# extract only the needed rows
new_header = tvt_p3.iloc[0] #grab the first row for the header
tvt_p3 = tvt_p3[2:]
tvt_p3.columns = newheader
```

```python
tvt_p3
```

```python
tvt_p3.reset_index(inplace =True)
```

```python
tvt_p3
```

```python
tvt_p3 =tvt_p3.rename (columns ={'index': "Month"})
```

```python
tvt_p3
```

```python
# Needed to conver the mumber to numeric becuse they are not numbers

Col_to_cov = ['Rural Interstate', 'Rural Other Arterial', 'Other Rural', 'Urban Interstate', 'Urban Other Arterial','Other Urban','All Systems']
```

```python
tvt_p3[Col_to_cov] = tvt_p3[Col_to_cov].apply(pd.to_numeric, errors= 'coerce')
```

```python
tvt_p3.set_index('Month', inplace= True)
```

```python
# Plot the graphs
tvt_p3[['Rural Interstate', 'Rural Other Arterial','Other Rural',
        'Urban Interstate','Urban Other Arterial', 'Other Urban']].plot(kind ='line', figsize= (10,6))

plt.title('Traffic Volume by Functional Class \n 2022')
plt.ylim=(0, 1)
plt.ylabel('Traffic Volumne in Billions')
plt.xlabel('Months')
plt.legend(title = 'Systems')
plt.text(-0.2, 4.0, 'Data source: FHWA', fontsize =6, color = 'gray')
plt.show()
```

## Processing page  4


```python
# Read file
tvt_p4 = pd.read_excel(file, sheet_name="Page 4", skiprows = 6, skipfooter= 2) # read file
```

```python
# View
tvt_p4
```

```python
# grab only needed column
tvt_p4 = tvt_p4[['Northeast','Unnamed: 7','Unnamed: 8']]


new_header = tvt_p4.iloc[0] #grab the first row for the header
tvt_p4 = tvt_p4[1:] #take the data less the header row
tvt_p4.columns = new_header #set the header row as the df header

# Replace nan with State as column name
tvt_p4= tvt_p4.rename(columns=str).rename(columns={'nan':'State'})


# filter out just the needed fileds
tvt_p4 = tvt_p4. loc[(tvt_p4['State'] != 'Subtotal') & (tvt_p4['State'] != 'South Atlantic')
                      & (tvt_p4['State'] != 'North Central') & (tvt_p4['State'] != 'South Gulf')
                      & (tvt_p4['State'] != 'West')]
```

```python
tvt_p4.set_index('State', inplace= True)
```

```python
tvt_p4
```

```python
tvt_p4[['PCMILES']].plot(kind ='bar', figsize= (10,6))

plt.title('Traffic Volume by State\n 2022')
plt.ylim=(0, 1)
plt.ylabel('Traffic Volumne in Billions')
plt.xlabel('State Name')
plt.legend(title = 'Traffic Volume')

plt.show()
```
