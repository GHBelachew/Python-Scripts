```python
# Summeize by municipal code
df_c12_Mun= df_c12_20_60_Tall.groupby(['County-Munic Code','Year'], as_index=False).agg({
    'Municipality':'first','REMI Region':'first','Total':'sum', 'Manuf': 'sum', 'Retail':'sum','Service':'sum','Other':'sum'})
```

```python
# Summeize by County code
df_c12_County= df_c12_20_60_Tall.groupby(['County Code','Year'], as_index=False).agg({
    'Total':'sum', 'Manuf': 'sum', 'Retail':'sum','Service':'sum','Other':'sum'})
```

```python
# Summeize by Remi Region
df_c12_RemiRegion= df_c12_20_60_Tall.groupby(['REMI Region','Year'], as_index=False).agg({
    'Total':'sum', 'Manuf': 'sum', 'Retail':'sum','Service':'sum','Other':'sum'})
```
